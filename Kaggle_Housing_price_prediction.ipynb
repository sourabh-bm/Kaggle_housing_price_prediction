```python
X =df.drop('price',axis =1).values #Values is used to convert the pandas dataframe into a numpy numeric array, 
#tensor flow is not good at working with pandas dataframe
```

```python
y =df['price'].values #Values is used to convert the pandas dataframe into a numpy numeric array, 
#tensor flow is not good at working with pandas dataframe
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scalar = MinMaxScaler()
```

```python
X_train = scalar.fit_transform(X_train)
```

```python
X_test =scalar.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()
```

```python
X_train.shape
```

```python
model.add(Dense(19,activation = 'relu'))
model.add(Dense(19,activation = 'relu'))
model.add(Dense(19,activation = 'relu'))
model.add(Dense(19,activation = 'relu'))

model.add(Dense(1))
model.compile(optimizer='adam',loss ='mse')
```

```python
model.fit(x= X_train,y =y_train,validation_data=(X_test,y_test),batch_size=128,epochs =400)#Validation data parameter does not improve the weight orbias computation of the neurons
# it is only to evaluate 'how well the model is performing'
```

```python
losses =pd.DataFrame(model.history.history)
losses.plot()
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,explained_variance_score
```

```python
predictions = model.predict(X_test)
```

```python
predictions.shape
```

```python
mean_squared_error(y_test,predictions)
```

```python
np.sqrt(mean_squared_error(y_test,predictions))
```

```python
mean_absolute_error(y_test,predictions)
```

```python
df['price'].describe()
```

```python
 5.402966e+05 #The mean of the price
```

```python
explained_variance_score(y_test,predictions)
```
